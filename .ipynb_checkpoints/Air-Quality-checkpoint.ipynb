> ## Functions

```python
def get_data(filepath):
    
    # Import data
    df = pd.read_csv('data/'+filepath, delimiter=';')
    
    return df


def group_data():
    # Create a list with all files in the folder named data
    files = glob('data/*')
    # Empty list which will receive the df wrangled
    frames = []
    for f in files:
        # To solve a error due to '\'
        f = f[5:]
        # Append into the frames list
        frames.append(get_data(f))
 
    # Concat the dfs into one named df
    df = pd.concat(frames)
       
    # Convert the timestamp type to datetime
    df['timestamp'] = pd.to_datetime(pd.Series(df['timestamp'])) 
    
    # Set timestamp index
    df = df.set_index('timestamp')

    # Localize reading time stamps to the timezone for "Africa/Nairobi".
    df.index = df.index.tz_convert('Africa/Nairobi')
    
    # Removing Outliers - mask
    mask = df['value'] < 350
    df = df[mask]
    
    # Resample the data to provide the mean value for each hour
    df = df['value'].resample('1H').mean().fillna(method='ffill')
    
    return df   
```

>> ## Time Series

```python
# The wrangled data will be stored into the variable y
y = group_data()
```

>> ## Plot the rolling average of the readings in y. Window size of 168 (the total of hours in a week).

>> ## Create an ACF plot for the data in y

- Measures the relationship between a time series and its own lagged version. Studying autocorrelation can provide valuable insights into the structure and patterns of a time series, allowing for more accurate predictions and the identification of possible trends and cycles.


- Anything inside the blue band the coefficient correlation is so small that it is not statistically significant. We are only interested in values really far from this band

- Apparently we have some good prediction power from 1 to 3 lag and some value near to 12 and 14 lags and 24 to 26. This gives an idea of how many lags we will include in our model, but the decision of how many lags include in our model comes from the PACF

```python
fig, ax = plt.subplots(figsize=(15,6))
plot_acf(y, ax=ax)
plt.xlabel("Lag [hours]")
plt.ylabel("Correlation Coefficient");
```

> # Auto Regressive (AR) Model
- Model used to analyze time series data. In an AR model, the current value of a variable is predicted based on past values of that variable.

- The basic idea behind an AR model is that the current value of a variable is a linear combination of its past values, with some added random error term.

>> ## Split

```python
# Define the size of the sample and the cutoff of the serie
cutoff = int(len(y) * .9)
y_train = y.iloc[:cutoff]
y_test = y.iloc[cutoff:]
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

>> ## Build Model

>>> ## Baseline Model
- It's a simple model that always returns the same value no matter the feature value. For example a model that always returns the mean of the distribution as a prediction of the target no matter the feature value.

```python
y_train_mean = y_train.mean()
y_pred_baseline = [y_train_mean] * len(y_train)
mae_baseline = mean_absolute_error(y_train, y_pred_baseline)
print(f"Mean P2 Reading: {y_train_mean}")
print(f"Baseline MAE: {mae_baseline}")
```

>> ## Iterate: Best Lag Parameter

- AR model to predict value readings, but which hyperparameter settings will give the best performance? Use a for loop to train your AR model on using settings for p from 1 to 30. Each time you train a new model, calculate its mean absolute error and append the result to the list. Then store your results in the Series mae_series.

```python
# Iterator object
p_params = range(1,31)
# Empty list to store the maes for each trained model using each p_params
maes = []
# Loop
for p in p_params:
    # Model AutoReg using the p as parameter for lag. This create null values equal to the number of lags
    model = AutoReg(y_train, lags=p).fit()
    # Predict. Drop the null values - Change the sample size
    y_pred = model.predict().dropna() 
    # Evaluate. The iloc[p:] is necessary to keep the sample size equivalent.
    mae = mean_absolute_error(y_train.iloc[p:], y_pred)
    # Append
    maes.append(mae)

# Create a Series based on maes values and the p_params as index
mae_series = pd.Series(maes, name='mae', index=p_params)
mae_series.head()
```

- Define the value of p that provides the best performance. The minimum mean absolute error shows the better model

>> ## Best Parameter

```python
best_p = mae_series[mae_series == min(mae_series)].index[0]
best_mae = min(mae_series)
print(f'The best model uses the lag of: {best_p} and return a mae of {best_mae}')
```

>>> ## Best Model

```python
best_model = AutoReg(y_train, best_p).fit()
```

### Residuals

```python
y_pred_model = best_model.predict().dropna()
y_train_resid = y_train.iloc[best_p:] - y_pred_model
y_train_resid.head()
```

```python
fig, ax = plt.subplots(figsize=(15,6))
y_train_resid.plot(kind='hist', ax=ax)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Best Model, Training Residuals');
```

>>> ## Best Model - Evaluate

```python
y_pred_test = best_model.predict(y_test.index.min(), y_test.index.max()) # I need to give two time stamps 
test_mae = mean_absolute_error(y_test, y_pred_test)
print("Test MAE:", test_mae)
```

- The result was not good. It was greater than y_train_pred

```python
df_pred_test = pd.DataFrame(
    {"y_test": y_test, "y_pred": y_pred_test}, index=y_test.index
)

fig, ax = plt.subplots(figsize=(15,6))
df_pred_test.plot( ax=ax)
plt.title('Real vs Predictions')
plt.show();
```

- Traditional test evaluation doesn't work well for time series. The predictions lose power over time because the model gets farther and farther away from its beginning

>>> ## Best Model - Evaluate: Walk-forward validation 

- Walk-forward Validation: Use only values from the train set. As we go forward we re-train the model at for each new observation in the dataset. It is a iterative process.


```python
# Create a Series to store the dayly predictions
y_pred_wfv = pd.Series()
# Begin with the y_train. This is the first step and on this we will be adding data as we continue making predictions
history = y_train.copy()
# for each value in the y_test
for i in range(len(y_test)):
    # train a model 
    model = AutoReg(history, lags=best_p).fit()
    # make a forecast for the next value
    next_pred = model.forecast()
    # Append the value forecasted into the Series    
    y_pred_wfv = pd.concat([y_pred_wfv, next_pred])
    # Expand the history (our train set) by one particular timestamp. Get the timestamp associated to the next_pred and append to the history
    history = pd.concat([history, y_test[next_pred.index]])
```

```python
test_mae = mean_absolute_error(y_test, y_pred_wfv)
print("Test MAE (walk forward validation):", round(test_mae, 2))
```
